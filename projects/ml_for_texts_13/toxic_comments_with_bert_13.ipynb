### Проведем подготовку признаков с помощью BERT

```python

model_name = "unitary/toxic-bert"

model = AutoModel.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

### Удалим из текста мусорные символы

```python
def clean_text(text):
    '''
    Функция оставляет все латинские буквы и пробелы
    '''
    text = re.sub(r'[^a-zA-Z ]',' ',text) # убираем лишние символы
    text_split = text.split() # избавляемся от ненужных пробелов
    text_joined = ' '.join(text_split)  # соединяем слова пробелами
#     print(text_joined)
    return text_joined
    
    
# clean_text('Explanation\nWhy the edits made under my usern')
```

```python
df['text'] = df['text'].apply(clean_text)
```

```python
df
```

### Токенизируем текст

```python
%%time
tokenized = df['text'].apply((lambda x: tokenizer.encode(x, padding=True, truncation=True,max_length=50, add_special_tokens = True)))
```

```python
max_len = 0
for i in tokenized.values:
    if len(i) > max_len:
        max_len = len(i)
```

```python
max_len
```

```python
%%time
padded = np.array([i + [0]*(max_len - len(i)) for i in tokenized.values])
```

```python
attention_mask = np.where(padded != 0, 1, 0)
```

```python
padded[1]
```

Теперь повторим эти действия на целой выборке

```python
%%time

batch_size = 400
embeddings = []



for i in tqdm(range(padded.shape[0]//batch_size)):
    batch = torch.LongTensor(padded[batch_size*i:batch_size*(i+1)]) 
    attention_mask_batch = torch.LongTensor(attention_mask[batch_size*i:batch_size*(i+1)])

#     print(batch.shape)
#     break
    with torch.no_grad():
        batch_embeddings = model(batch, attention_mask=attention_mask_batch)

    embeddings.append(batch_embeddings[0][:,0,:].numpy())
```

Это время выполнения в коллабе



А это время выполнения в локальном юпитере



```python
features = np.concatenate(embeddings)
target = df['toxic']
```

```python
print(features.shape)
print(target.shape)
```

```python
target = target[:features.shape[0]]
print(target.shape)
```

```python
features_train, features_test, target_train, target_test = train_test_split(
                features, target, test_size=0.2, random_state=12345)
```

### Сделаем обучающую и тестовую выборки

```python
features_train, features_test, target_train, target_test = train_test_split(
                features, target, test_size=0.2, random_state=12345)
```

### Логистическая регрессия 

```python

pipe_l = Pipeline([
    (('model', LogisticRegression(random_state=12345, solver='liblinear')))
    ])

param_grid_l = [
        {   'model': [LogisticRegression(random_state=12345, solver='liblinear')],
             'model__C': np.arange(.1,1.,.3),
            'model__penalty':['l1', 'l2'],
            'model__max_iter': range(20, 400, 100)
        }]
grid_l = GridSearchCV(pipe_l, param_grid=param_grid_l, scoring='f1', cv=kf, verbose=True)
```

```python
best_grid_l = grid_l.fit(features_train, target_train)
print('Best parameters is:', grid_l.best_params_)
print('Best score is:', grid_l.best_score_)
```

### CatBoost

#### Обучение CatBoost на выборке с дисбалансом

```python

pipe_cat = Pipeline([
    (('model', CatBoostClassifier(random_state=12345)))
    ])

param_grid_cat = [
        {   'model': [CatBoostClassifier(random_state=12345)],
             'model__iterations': np.arange(50,201,50),
            'model__learning_rate':np.arange(0.1,0.3,0.1)
        }]
grid_cat = GridSearchCV(pipe_cat, param_grid=param_grid_cat, scoring='f1', cv=kf, verbose=True)


best_grid_cat = grid_cat.fit(features_train, target_train)
print('Best parameters is:', grid_cat.best_params_)
print('Best score is:', grid_cat.best_score_)
```
