### –§—É–Ω–∫—Ü–∏—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è —Ñ–∏—á–µ–π

```python
def make_features(df, max_lag, rolling_size):
    df_copy = df.copy()
    
    cols = df_copy.columns[1:]
    df_copy = df_copy.drop(cols, axis=1)
#     print(df)
    
    # –¥–µ–ª–∞–µ–º –∫–∞–ª–µ–Ω–¥–∞—Ä–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
#     df['year'] = df.index.year —ç—Ç–∞ —Ñ–∏—á–∞ –Ω–µ –∏–º–µ–µ—Ç —Å–º—ã—Å–ª–∞
    df_copy['month']= df.index.month
    df_copy['day'] = df.index.day
    df_copy['dayofweek'] = df.index.dayofweek
    df_copy['hour'] = df.index.hour
    df_copy['weekend'] = df_copy['dayofweek'] > 5
    df_copy['am'] = df_copy['hour'] < 12
    
    
    # –¥–µ–ª–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –æ—Ç—Å—Ç–∞—é—â–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π
    for lag in range(1, max_lag+1):
        df_copy[f'lag_{str(lag)}'] = df_copy['num_orders'].shift(lag)
        
    # –¥–µ–ª–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫ —Å–∫–æ–ª—å–∑—è—â–µ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —Ä–∞–∑–º–µ—Ä–æ–º "rolling size"
    df_copy['rolling_mean'] = df_copy['num_orders'].shift().rolling(rolling_size).mean()
    
    # –î–æ–±–∞–≤–ª—è–µ–º —Ä–∞–∑–Ω–∏—Ü—É, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç —Ü–µ–ª–µ–≤—ã–º –ø—Ä–∏–∑–Ω–∞–∫–æ–º
#     df['dif'] = df['num_orders'] - df['num_orders'].shift()
    
    return df_copy
```
